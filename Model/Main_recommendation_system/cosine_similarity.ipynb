```python
beer_df = pd.read_csv('data-files/sample.csv')
```

```python
# Cosine Similarity를 계산할 부분 추출
taste_mat = beer_df[['sweet','sour','tender','bitter','abundant']]

# Cosine Similarity 계산
from sklearn.metrics.pairwise import cosine_similarity

taste_sim = cosine_similarity(taste_mat, taste_mat)
print(taste_sim.shape)
print(taste_sim[:2])
```

```python
# Cosine Similarity를 계산한 ndarray 파일로 저장
np.save('outputs/taste_sim.npy', taste_sim)
```
